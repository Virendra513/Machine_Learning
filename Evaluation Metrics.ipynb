
# Model evaluation
it is a crucial step in the machine learning workflow to assess the performance of a trained model on unseen data. Various metrics and techniques are used depending on the nature of the problem (classification, regression, etc.). Here are some common methods for model evaluation:

## Classification Matrics
* confusion_matrix 
* accuracy_score
* precision_score
* f1_score
* recall_score
* roc_auc_score
* log_loss
* classification_report

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score,recall_score, roc_auc_score, log_loss, classification_report
```

```python
# load Iris datset for classifcation
iris= load_iris()
x,y =iris.data, iris.target
print(x)
print(y)
```

```python
x.shape, y.shape
```

```python
# Split the dataset
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=42)

# Train a Random Forest Classifier
clf= RandomForestClassifier(random_state=42)
clf.fit(x_train,y_train)

# Make prediction
y_pred=clf.predict(x_test)
y_prob= clf.predict_proba(x_test)

# Evaluate the model
print("Confusion Matrix")
print(confusion_matrix(y_test,y_pred))
```

```python
print('\nAccuracy',accuracy_score(y_test, y_pred))
```

```python
print("\n Precision",precision_score(y_test, y_pred, average='macro'))
```

```python
print("\n Recall",recall_score(y_test, y_pred, average='macro'))
```

```python
print("f1_score",f1_score(y_test,y_pred, average='macro'))
```

```python
print("ROC AUC Score:", roc_auc_score(y_test, y_prob, multi_class='ovr'))
```

```python
print("Log Loss:", log_loss(y_test, y_prob))
```

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

# Regression Example

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.datasets import fetch_california_housing
```

```python
# Load California_housing
housing = fetch_california_housing()
x,y= housing.data, housing.target
```

```python
# Split the dataset
`_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# Train a Random Forest Regressor
reg= RandomForestRegressor()
reg.fit(x_train, y_train)
```

```python
 # Make predictions
y_pred= reg.predict(x_test)
```

```python
# Evaluate the model
print("Mean Absoluter Error :",mean_absolute_error(y_test, y_pred))
print("Mean Absoluter Error :",mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", mean_squared_error(y_test, y_pred, squared=False))
print("R- squared", r2_score(y_test, y_pred))
```


## R-squared
also known as the coefficient of determination, is a statistical measure that represents the proportion of the variance in the dependent variable that is predictable from the independent variables in a regression model. It ranges from 0 to 1, where:


R^2=0: The model explains none of the variability of the response data around its mean.

R^2=1: The model explains all the variability of the response data around its mean.

The formula for R-squared is:
R^2=1(SSR/SST)
where:

SSR (Sum of Squared Residuals) is the sum of the squared differences between the actual values (yi-yi^hat)
SST (Total Sum of Squares) is the sum of the squared differences between the actual values (yi-yi^-) and the yi^- is the mean of the dependent variable 
yi^hat
